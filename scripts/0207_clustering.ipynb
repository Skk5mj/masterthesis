```python
results_list = [
    "../log/model1",
    "../log/model2",
    "../log/model3",
    "../log/model4",
]
k = 7
for i, n in enumerate(results_list):
    result_path = osp.join(n, "results.pkl")
    print(n)
    with open(result_path, 'rb') as p:
        l = pickle.load(p)
    print(l["params"])
    print(np.mean(l["mae"]), np.std(l["mae"]))
    print(l["best_fold"])
    model = Network3layers(indim = 116, ratio = 0.5, k = k + i, nclass = 1, R = 116)
    best_weights = osp.join(n, f"fold_{l['best_fold']}/model_weights.pth")
    cluster =  plot_cluster_AAL(model, best_weights, model_number=i + 1)
    out = osp.join(n, f"model{i+1}clustering_results.pkl")
    with open(out, 'wb') as p:
        pickle.dump(cluster, p)
```
